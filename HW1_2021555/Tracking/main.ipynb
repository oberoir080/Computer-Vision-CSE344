```python
import numpy as np 
import pandas as pd 
import cv2
import os
import matplotlib.pyplot as plt
```

```python
PATH = "D:\\Git\\CV\\HW1_2021555\\Tracking\\data\\train\\"
```

```python
frame = cv2.cvtColor(cv2.imread(PATH + "MOT17-11-SDP\\img1\\000001.jpg"), cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 8))
plt.imshow(frame)
plt.axis('off')
plt.show()
```

```python
frame_count = {}

for i in os.listdir(PATH):
    frame_count[i] = len(os.listdir(PATH + i + "\\img1\\"))

plt.figure(figsize=(10, 8))
plt.bar(frame_count.keys(), frame_count.values())
plt.xlabel('Sequences')
plt.ylabel('Number of Frames')
plt.title('Data Distribution by Number of Frames')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
